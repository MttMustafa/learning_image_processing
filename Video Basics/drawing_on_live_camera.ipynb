```python
#drawing a static rectangle on a video

import cv2
capture = cv2.VideoCapture(0)

width = int(capture.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

#top left corner of the rectangle
x = width // 2 # double slash '//' operator always returns integer 
y = height // 2
#width and heght of the rectangle
w = width // 4
h = height // 4

#bottom right corner of the rectangle will be w+w, y+h


while True:
    ret, frame = capture.read()
    
    cv2.rectangle(frame, (x, y), (x+w, y+h), color = (0, 255, 0), thickness = 4)
    
    cv2.imshow('frame', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
capture.release()
cv2.destroyAllWindows()
```

```python
#drawing a dynamic rectangle on a video with mouse click

import cv2

#callback function for setMouseCallBack
def draw_rectangle(event, x, y, flags, param):
    
    global pnt_1, pnt_2, topLeft_clicked, botRight_clicked
    
    #if user clicked on screen twice and caused rectangle to be drawn
    if event == cv2.EVENT_LBUTTONDOWN:
        if topLeft_clicked == True and botRight_clicked == True:
            pnt_1 = (0, 0)
            pnt_2 = (0, 0)
            topLeft_clicked = False
            botRight_clicked = False
        #if user did not clicked on screen
        if topLeft_clicked == False:
            pnt_1 = (x, y)
            topLeft_clicked = True
        #if user clicked on screen once but not the second time
        elif botRight_clicked == False:
            pnt_2 = (x, y)
            botRight_clicked = True

pnt_1 = (0, 0)
pnt_2 = (0, 0)
topLeft_clicked = False
botRight_clicked = False

capture = cv2.VideoCapture(0)

cv2.namedWindow('Video_1')
cv2.setMouseCallback('Video_1', draw_rectangle)

while True:
    ret, frame = capture.read()
    
    if topLeft_clicked:
        cv2.circle(frame, center = pnt_1, radius = 5, color = (0, 0, 255), thickness = -1)
    if topLeft_clicked and botRight_clicked:
        cv2.rectangle(frame, pnt_1, pnt_2, (0, 0, 255), 3)
    
    cv2.imshow('Video_1', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
capture.release()
cv2.destroyAllWindows()
```

```python
#drawing a dynamic rectangle on a video with mouse click

import cv2

#callback function for setMouseCallBack
def draw_circle(event, x, y, flags, param):
    
    global pnt, clicked
    

    if event == cv2.EVENT_RBUTTONDOWN:
        if clicked == True:
            pnt = (0, 0)
            clicked = False
        #if user did not clicked on screen
        else:
            pnt = (x, y)
            clicked = True


pnt = (0, 0)
clicked = False

capture = cv2.VideoCapture(0)

cv2.namedWindow('Video_1')
cv2.setMouseCallback('Video_1', draw_circle)

while True:
    ret, frame = capture.read()
    
    if clicked:
        cv2.circle(frame, center = pnt, radius = 200, color = (255, 0, 0), thickness = 2)

    cv2.imshow('Video_1', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
capture.release()
cv2.destroyAllWindows()
```
