```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("D:\\YEDEK\\Belgeler\\WS\\Python\\Computer Vision Course\Computer-Vision-with-Python\\Computer-Vision-with-Python\\DATA\\internal_external.png", 0)
```

```python
plt.imshow(img, cmap = "gray")
```

```python
#contour locs and hierarchies
contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours
```

```python
hierarchy
```

```python
external_contours = np.zeros(img.shape)
```

```python
hierarchy.shape
```

```python
for i in range(len(contours)):
    #checks external countours -1 means external other number means internal
    if hierarchy[0][i][3] == -1:
        cv2.drawContours(external_contours, contours, i, 255, -1)
plt.imshow(external_contours, cmap = "gray")
```

```python
internal_contours = np.zeros(img.shape)

for i in range(len(contours)):
    #checks internal contours
    #internal contours grouped in different numbers
    #for example smiley face internal contours grouped as 0
    #pizzas grouped as 4 
    #(checked the hierarchy matrix 3rd column)
    if hierarchy[0][i][3] != -1:
        cv2.drawContours(internal_contours, contours, i, 255, -1)
plt.imshow(internal_contours, cmap = "gray")
```
