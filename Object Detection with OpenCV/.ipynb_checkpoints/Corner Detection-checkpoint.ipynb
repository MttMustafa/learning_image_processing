```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
flat_chess = cv2.imread("D:\\YEDEK\\Belgeler\\WS\\Python\\Computer Vision Course\Computer-Vision-with-Python\\Computer-Vision-with-Python\\DATA\\flat_chessboard.png")
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)
plt.imshow(flat_chess)
```

```python
gray_flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_flat_chess, cmap = 'gray')
```

```python
real_chess = cv2.imread("D:\\YEDEK\\Belgeler\\WS\\Python\\Computer Vision Course\Computer-Vision-with-Python\\Computer-Vision-with-Python\\DATA\\real_chessboard.jpg")
gray_real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_real_chess, cmap = 'gray')
```

```python
gray = np.float32(gray_flat_chess)
#dst algorithm maps the images in a way to emphasize corners
#                                              ##kernel size##kernel default
dst = cv2.cornerHarris(src = gray, blockSize = 2, ksize = 3, k = 0.04)
plt.imshow(dst, cmap = 'gray')
```

```python
dst = cv2.dilate(dst, None) #morphological operator
plt.imshow(dst, cmap = 'gray')
```

```python
flat_chess_copy = flat_chess.copy()
#dst values that greater that 1% of the max value will be drawn red
flat_chess_copy[dst > 0.01 * dst.max()] = [255,0,0]
plt.imshow(flat_chess_copy)
```

```python
gray_real = np.float32(gray_real_chess)
#                                              ##kernel size##kernel default
dst_real = cv2.cornerHarris(src = gray_real, blockSize = 2, ksize = 3, k = 0.04)
plt.imshow(dst_real, cmap = 'gray')
```

```python
dst_real = cv2.dilate(dst_real, None)
plt.imshow(dst_real, cmap = 'gray')
```

```python
real_chess_rgb = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)
real_chess_rgb[dst_real > 0.01 * dst_real.max()] = [255, 0, 0]
plt.imshow(real_chess_rgb)
```

```python
#SHI-TOMASI CORNER DETECTION ALGORITHM
```

```python
#                               #max num of corners,#quality level
corners = cv2.goodFeaturesToTrack(gray_flat_chess,5, 0.01, 10)
```

```python
corners = np.int0(corners)
```

```python
for i in corners:
    x, y = i.ravel()
    cv2.circle(flat_chess, (x, y), 3, (255,0,0), -1)
plt.imshow(flat_chess)
```

```python
corners_2 = cv2.goodFeaturesToTrack(gray_real_chess,80, 0.01, 10)
corners_2 = np.int0(corners_2)
for i in corners_2:
    x, y = i.ravel()
    cv2.circle(real_chess, (x, y), 3, (255,0,0), -1)
plt.imshow(real_chess)
```
