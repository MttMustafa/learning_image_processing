```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
face = cv2.imread('D:\YEDEK\Belgeler\WS\Python\Computer Vision Course\Computer-Vision-with-Python\Computer-Vision-with-Python\DATA\sammy_face.jpg')
plt.imshow(face)
```

```python
plt.imshow(cv2.Canny(image = face, threshold1 = 100, threshold2 = 200))
```

```python
median_val = np.median(face)
lower_t = int(max(0, 0.7 * median_val)) #lower threshold is either 0 or 70% of median whichever is smaller
upper_t = int(min(255, 1.3 * median_val)) #upper threshold is either 255 or 130% whichever is bigger
```

```python
plt.imshow(cv2.Canny(image = face, threshold1 = lower_t, threshold2 = upper_t))
```

```python
blurred_face = cv2.blur(face, ksize = (7,7))
```

```python
edges = cv2.Canny(image = blurred_face, threshold1 = lower_t, threshold2 = upper_t + 45)
plt.imshow(edges)
```
