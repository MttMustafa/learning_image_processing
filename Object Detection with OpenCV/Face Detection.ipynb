```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize= (12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
face_cascade = cv2.CascadeClassifier("D:\\YEDEK\\Belgeler\\WS\\Python\\Computer Vision Course\Computer-Vision-with-Python\\Computer-Vision-with-Python\\DATA\\haarcascades\\haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):

    face_img = img.copy()

    face_rects = face_cascade.detectMultiScale(face_img, scaleFactor = 1.2, minNeighbors = 5)

    for(x, y, w, h) in face_rects:
        cv2.rectangle(face_img, (x, y), (x + w, y + h), (255, 255, 255), 10)
    
    return face_img
```

```python
result = detect_face(solvay)
```

```python
display(result)
```

```python
eye_cascade = cv2.CascadeClassifier("D:\\YEDEK\\Belgeler\\WS\\Python\\Computer Vision Course\Computer-Vision-with-Python\\Computer-Vision-with-Python\\DATA\\haarcascades\\haarcascade_eye.xml")
```

```python
def detect_eye(img):

    eye_img = img.copy()

    eye_rects = eye_cascade.detectMultiScale(eye_img, scaleFactor = 1.2, minNeighbors = 5)

    for(x, y, w, h) in eye_rects:
        cv2.rectangle(eye_img, (x, y), (x + w, y + h), (255, 255, 255), 10)
    
    return eye_img
```

```python
result = detect_eye(nadia)
display(result)
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read(0)
    frame = detect_face(frame)
    cv2.imshow('Face Detection', frame)

    k = cv2.waitKey(1)

    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
