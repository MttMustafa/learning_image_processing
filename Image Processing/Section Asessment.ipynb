```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_img(img):
    fig= plt.figure(figsize = (15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
giraffes = cv2.imread('../../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/giraffes.jpg')
giraffesRGB = cv2.cvtColor(giraffes, cv2.COLOR_BGR2RGB)
```

```python
show_img(giraffesRGB)
```
