```python
#Pasting a smaller image over bigger one
img2 = cv2.imread('../../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.png')
img3 = cv2.imread('../../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
```

```python
x_start = img2.shape[1] - 300
y_start = img2.shape[0] - 300
```

```python
#Blending images with different sizes
#Pasting a smaller image over bigger one
img2 = cv2.imread('../../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/dog_backpack.png')
img3 = cv2.imread('../../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/watermark_no_copy.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img3 = cv2.resize(img3,(300,300))
```

```python
regionOfInterest = img2[y_start:, x_start:]
plt.imshow(regionOfInterest)
```

```python
img3gray = cv2.cvtColor(img3, cv2.COLOR_RGB2GRAY)
plt.imshow(img3gray, cmap = 'gray')
```

```python
mask_inverse = cv2.bitwise_not(img3gray)
plt.imshow(mask_inverse, cmap = 'gray')
```

```python
white_background = np.full(img3.shape, 255, dtype = np.uint8)
```

```python
reversedImgWithColorSpace = cv2.bitwise_or(white_background, white_background, mask = mask_inverse)
plt.imshow(reversedImgWithColorSpace)
```

```python
originalImgWithBlankBackground = cv2.bitwise_or(img3, img3, mask = mask_inv)
plt.imshow(originalImgWithBlankBackground)
```

```python
final_roi = cv2.bitwise_or(regionOfInterest, originalImgWithBlankBackground)
```

```python
plt.imshow(final_roi)
```

```python
img2[y_start:, x_start:] = final_roi
plt.imshow(img2)
```
