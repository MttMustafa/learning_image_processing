```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img, text= "TEXT", org = (120,300), fontFace = font, fontScale = 5, color = (255, 255, 255), thickness = 20,)
    return blank_img
```

```python
def show_img(img):
    fig= plt.figure(figsize = (15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
show_img(load_img())
```

```python
kernel = np.ones((5,5), dtype = np.uint8)
```

```python
kernel
```

```python
img1 = load_img()
```

```python
result1 = cv2.erode(img1, kernel, iterations = 1)
```

```python
show_img(result1)
```

```python
result2 = cv2.erode(img1, kernel, iterations = 3)
show_img(result2)
```

```python
img2 = load_img()
```

```python
white_noise = np.random.randint(low = 0, high = 2, size = (600, 600))
```

```python
white_noise
```

```python
show_img(white_noise)
```

```python
white_noise *= 255
```

```python
white_noise
```

```python
img2 += white_noise
```

```python
show_img(img2)
```

```python
deNoising  = cv2.morphologyEx(img2, cv2.MORPH_OPEN, kernel)
```

```python
show_img(deNoising)
```
