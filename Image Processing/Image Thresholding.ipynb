```python
img2 = cv2.imread('../../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/crossword.jpg', 0)
```

```python
plt.imshow(img2, cmap='gray')
```

```python
def show_pic(img):
    fig= plt.figure(figsize = (15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
show_pic(img2)
```

```python
nret, nthr = cv2.threshold(img2, 160, 255, cv2.THRESH_BINARY)
show_pic(nthr)
```

```python
nthr2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,8) #number of neightbor pixels, substract constant
show_pic(nthr2)
```

```python
blended = cv2.addWeighted(src1 = nthr, alpha = 0.5, src2 = nthr2, beta = 0.5, gamma = 0)
show_pic(blended)
```
