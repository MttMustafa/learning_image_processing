```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def show_img(img):
    fig= plt.figure(figsize = (15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap = 'gray')
```

```python
img1 = cv2.imread('../../Computer-Vision-with-Python/Computer-Vision-with-Python/DATA/sudoku.jpg', 0)
```

```python
show_img(img1)
```

```python
sobelx = cv2.Sobel(img1, cv2.CV_64F, 1, 0, ksize = 5) #horizontal derivative of the image
```

```python
show_img(sobelx)
```

```python
sobely = cv2.Sobel(img1, cv2.CV_64F, 0, 1, ksize = 5) #vertical derivative of the image
```

```python
show_img(sobely)
```

```python
laplacian1 = cv2.Laplacian(img1, cv2.CV_64F)
show_img(laplacian1)
```

```python
blended1 = cv2.addWeighted(src1 = sobelx, alpha = 0.5, src2 = sobely, beta = 0.5, gamma = 0)
show_img(blended1)
```

```python
ret, th1 = cv2.threshold(img1, 100, 255, cv2.THRESH_BINARY)
show_img(th1)
```

```python
kernel = np.ones((4,4), np.uint8)
gradient1 = cv2.morphologyEx(blended1, cv2.MORPH_GRADIENT, kernel)
```

```python
show_img(gradient1)
```
